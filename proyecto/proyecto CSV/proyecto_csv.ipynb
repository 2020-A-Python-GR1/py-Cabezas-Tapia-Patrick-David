### Cargar el Dataset CSV

```python
path_data = '/home/d/Documentos/programacion/py-cabezas-tapia-patrick-david/proyecto/proyecto CSV/MoviesOnStreamingPlatforms_updated.csv'
df = pd.read_csv(path_data)
df
```

## Graficas

### 1. Gráficas de los años en el que se lanzo la película.

```python
df['Age'].value_counts().plot(kind='pie',cmap ='Paired')
```

### 2. Gráficas de los países donde se lanza las películas (30 principales)

```python
df['Country'].value_counts()[:30].plot(kind='bar',legend='Reverse')
plt.xlabel('Country')
plt.ylabel('Quantity')
```

### 3. Gráfica de los directores con mas títulos publicados. (30 principales)

```python
df['Directors'].value_counts()[:30].plot(kind='barh',legend='Reverse')
plt.ylabel('Directors')
plt.xlabel('Quantity')
```

### 4. Gráfica de los años en la que se lanzaron mas películas

```python
df['Year'].value_counts()[:30].plot(kind='bar')
plt.xlabel('Year')
plt.ylabel('Quantity')
```

### 5. Gráfica las películas mejor puntuadas en IMDB. (50 principales)

```python
df_columnas = ['Title','IMDb']
valores = df[df_columnas].sort_values(by='IMDb', ascending = False).head(50)
valores.index = valores['Title']
valores.plot( kind='bar',figsize=(17,5))
```

### 6. Gráfica las películas peor puntuadas en IMDB. (50 principales)

```python
df_columnas = ['Title','IMDb']
valores = df[df_columnas].sort_values(by='IMDb', ascending = True).head(50)
valores.index = valores['Title']
valores.plot( kind='bar',figsize=(17,5))
```

### 7. Gráfica de las películas puntuadas al 100% con los directores mas repetidos. 

```python
valores = df[df['Rotten Tomatoes']=='100%']['Directors'].value_counts()[df[df['Rotten Tomatoes']=='100%']['Directors'].value_counts()>1]
valores.plot( kind='bar',figsize=(17,5))
plt.xlabel('Directors')
plt.ylabel('Quantity')
```

### 8. Gráfica de los géneros que mas se repiten en las películas

```python
df['Genres'].value_counts()[:30].plot(kind='barh',legend='Reverse',figsize=(12,15), color='#00ff00')
plt.ylabel('Directors')
plt.xlabel('Quantity')
```

### 9. Grafica de los lenguajes mas populares

```python
df['Language'].value_counts().head(30).plot(kind='bar',figsize=(17,6), color='#00fff0')
```

### 10. Grafica de la duraicon promedio de las peliculas

```python
df['Runtime'].value_counts().head(30).plot(kind='bar',figsize=(17,6), color='#0ffff0')
```

### 11. Grafica de las peliculas con mayor duacion en minutos

```python
df_columnas = ['Title','Runtime']
valores = df[df_columnas].sort_values(by='Runtime', ascending = False).head(50)
valores.index = valores['Title']
valores.plot( kind='bar',figsize=(17,5), color='#0ffff0')
```
