```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
path_data = '/home/d/Documentos/programacion/py-cabezas-tapia-patrick-david/proyecto/proyecto scrapy/proyecto_scrapy/lista.csv'
df = pd.read_csv(path_data)
df
```

### 1. Gráficas de las Uploader que han cargado archivos 

```python
df['Uploader'].value_counts().plot(kind='pie',cmap ='Paired')
```

### 2. Gráficas de las Year que mas achivos subieron

```python
df['Year'].value_counts().plot(kind='bar',legend='Reverse')
```

### 3. Gráfica de los titulos con mayor Rating (30 principales)

```python
df_columnas = ['Titulo','Rating']
valores = df[df_columnas].sort_values(by='Rating', ascending = False).head(30)
valores.index = valores['Titulo']
valores.plot( kind='bar',figsize=(17,5))
```

### 4. Gráfica de los titulos con menor Rating (30 principales)

```python
df_columnas = ['Titulo','Rating']
valores = df[df_columnas].sort_values(by='Rating', ascending = True).head(30)
valores.index = valores['Titulo']
valores.plot( kind='bar',figsize=(17,5))
```

### 5. Gráfica de los titulos con que han sido mas vistos por la comunidad (30 principales)

```python
df_columnas = ['Titulo','vistas']
valores = df[df_columnas].sort_values(by='vistas', ascending = False).head(30)
valores.index = valores['Titulo']
valores.plot( kind='bar',figsize=(17,5))
```

### 6. Gráfica de los titulos con que a la comunidad no le interesan (30 principales)

```python
df_columnas = ['Titulo','vistas']
valores = df[df_columnas].sort_values(by='vistas', ascending = True).head(30)
valores.index = valores['Titulo']
valores.plot( kind='bar',figsize=(17,5))
```

### 7. Gráfica de los titulos con mayor Rating y con muchas visitas

```python
df_columnas = ['Titulo','Rating']
valores = df[df_columnas][df['Rating']>4][df['vistas']>80000].sort_values(by='Rating', ascending = False)
valores.plot( kind='bar',figsize=(17,5))
```

### 8. Gráfica de los titulos con mayor numeros de votos de la comunidad (30 principales)

```python
df_columnas = ['Titulo','Votos']
valores = df[df_columnas].sort_values(by='Votos', ascending = False).head(30)
valores.index = valores['Titulo']
valores.plot( kind='bar',figsize=(17,5))
```

### 9. Gráfica de los titulos con menor numeros de votos de la comunidad (30 principales)

```python
df_columnas = ['Titulo','Votos']
valores = df[df_columnas].sort_values(by='Votos', ascending = True).head(30)
valores.index = valores['Titulo']
valores.plot( kind='bar',figsize=(17,5))
```

### 10. Gráfica de los titulos con menor numeros de votos de la comunidad con mayor Rating

```python
df_columnas = ['Titulo','Rating']
valores = df[df_columnas][df['Rating']>4][df['Votos']>100].sort_values(by='Rating', ascending = False)
valores.plot( kind='bar',figsize=(17,5))
```

### 11. Gráfica de los titulos con mayor Rating del 2018 

```python
df_columnas = ['Rating','Titulo']
df[df_columnas][df['Year']==2018.0][df['Rating']>4].plot(kind='bar',legend='Reverse',figsize=(17,5))
```

### 12. Gráfica de los Year con mayo cantidad de titulos con mejor Rating 

```python
df['Year'][df['Rating']>4].value_counts().plot( kind='bar',figsize=(17,5))
```

### 13. Gráfica de la distribucion de Uploader (30 principales)

```python
df['Uploader'].value_counts().plot( kind='bar',figsize=(17,5))
```

### 14. Gráfica de los Uploader con Rating mayor a 4.4

```python
df['Uploader'][df['Rating']>4.4].value_counts().plot( kind='bar',figsize=(17,5))
```

### 15. Gráfica de los titulos de Uploader con mayor Rating de ElAmigos

### 16. Gráfica de los titulos de Uploader con mayor Rating de CODEX

### 17. Gráfica de los titulos con mas vistas del 2018
